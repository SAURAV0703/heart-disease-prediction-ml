```python
import pandas as pd

df = pd.read_csv("/content/cardio_train.csv", sep=";")

df.head()
```

```python
df.info()

print("Missing Values:\n", df.isnull().sum())

print("Duplicate Rows:", df.duplicated().sum())
```

```python
df['age_years'] = (df['age'] / 365).astype(int)  # Convert to whole years
df.head()
```

```python
df = df.drop(columns=['age'])
df.head()
```

```python
df['bmi'] = df['weight'] / (df['height'] / 100) ** 2  # Convert height to meters and apply BMI formula
df.head()
```

```python
df['bmi'].describe()
```

```python
df_outliers = df[(df['bmi'] < 10) | (df['bmi'] > 50)]
print("Outliers Count:", df_outliers.shape[0])
df_outliers.head()
```

```python
df = df[(df['bmi'] >= 10) & (df['bmi'] <= 50)]
df.shape  # Check new dataset size
```

```python
df = df.sample(n=5000, random_state=42)  # Select 5,000 random rows
df.shape  # Check new dataset size
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
cols_to_scale = ['height', 'weight', 'ap_hi', 'ap_lo', 'bmi']
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])

df.head()  # Check scaled values
```

```python
df.to_csv("cleaned_heart_disease_data.csv", index=False)
```
